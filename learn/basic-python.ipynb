```python
# create readonly constants variable, in python 
# we can not create a constant variable but we can use a 
# convention to indicate that a variable should not be changed
PI = 3.14
```

```python
# or we can create Custom class for constants (read-only variables)
class Constants:
    def __init__(self):
        self._PI = 3.14
        self._GRAVITY = 9.8
    @property
    def PI(self):
        return self._PI
    @property
    def GRAVITY(self):
        return self._GRAVITY

constants = Constants()
print(constants.PI)
print(constants.GRAVITY)
# constants.PI = 3.15  # Uncommenting this will raise an error
```

```python
 
class Car:
    def __init__(self, make: int, model: str):
        self._make = make
        self._model = model
        
    @property
    def make(self):
        print(f"Getting make: {self._make}")
        return self._make
    
    @make.setter
    def make(self, value):
        print(f"Setting make {self._make} to value: {value}")
        self._make = value
         
        
car = Car(2020, "Toyota")
car.make = 2021
print(car.make) 
```

```python

from abc import ABC, abstractmethod


class Animal(ABC):
    
    
    @abstractmethod
    def sound(self):
        print("Animal sound")
 
class Dog(Animal):
    def __init__(self, name):
        self.name = name
    def sound(self):
        print(f"{self.name} sound: Woof Woof")
dog = Dog("Buddy")
dog.sound()
```
